# LEVEL 1 : 

### TASK 1 : Top Cuisines

#### Goal: Calculate the percentage of restaurants that serve each of the top cuisines.

We’ll break this into 3 stages:

1 . Data transformation (prepare data)

2 . Calculation (counts & percentages)

3 . Visualization (bar chart)

### STAGE 1 : DATA TRANSFORMATION 

```python
cuisines_series = df['cuisines'].astype(str).str.split(',').explode().str.strip()
```

1 . .astype(str)

* Converts all values to string type — ensures there are no missing values (NaN) or numeric errors.

Example:
* If some cells are blank or numbers, this step avoids errors like “.str method not supported for float”.

2 . .str.split(',')

Splits each string wherever a comma appears, turning it into a list of cuisines.

3 . .explode()

This is the magic step ✨

It creates a new row for each value in the list.

4 . str.strip()

Removes any extra spaces around the cuisine names.

Example: " North Indian " → "North Indian"

This ensures your data is clean and consistent before counting

##### Why do we use this : 
Because many restaurants have multiple cuisines in one cell, to find which cuisines are most common,
We need to separate them into individual rows before counting.

### STAGE 2 : COUNTING AND CALCULATING PERCENTAGE 

```python
cuisine_counts = cuisines_series.value_counts()
cuisine_counts
```

* value_counts() → Counts how many times each unique value appears in the Series.

* This gives you the number of restaurants serving each cuisine.

```python
# Calculate the percentage of total restaurants 

cuisine_percent = (cuisine_counts / len(df)) * 100
cuisine_percent
```

Explanation : 

* len(df) → gives total number of restaurants (rows in dataset).

* You divide each cuisine count by this total to get a proportion.

* Multiply by 100 → converts proportion to percentage.

✅ Example:
* If total restaurants = 2000 and North Indian = 400,
then (400 / 2000) * 100 = 20%.

* So 20% of restaurants serve North Indian cuisine.

### STAGE 3 : CREATING A CLEAN SUMMARY TABLE 

```python
top_cuisines = pd.DataFrame({
    'Cuisine': cuisine_counts.index,
    'Count': cuisine_counts.values,
    'Percentage': cuisine_percent.round(2).values
})
```

| Column         | Data                              | Explanation                             |
| -------------- | --------------------------------- | --------------------------------------- |
| `'Cuisine'`    | `cuisine_counts.index`            | Cuisine names (like “North Indian”)     |
| `'Count'`      | `cuisine_counts.values`           | Number of restaurants serving it        |
| `'Percentage'` | `cuisine_percent.round(2).values` | Rounded percentage (two decimal places) |


Purpose: This gives you a clean, tabular summary ready to print or plot.

```python
# --- Show only the Top 10 cuisines ---
top10_cuisines = top_cuisines.head(10)
print("Top 10 Cuisines by Restaurant Count and Percentage:\n")
print(top10_cuisines)
```

### STEP 4 : VISUALISATION (BAR CHARTS)

```python
plt.figure(figsize=(9,5))
plt.barh(top10_cuisines['Cuisine'], top10_cuisines['Percentage'], color='skyblue')
plt.title("Top 10 Cuisines by Percentage of Restaurants")
plt.xlabel("Percentage of Restaurants (%)")
plt.ylabel("Cuisine Type")
plt.gca().invert_yaxis()  # Highest at top
plt.show()
```

#### TASK 1 : FINAL ANALYSIS 

“The top cuisines served across restaurants are North Indian (41.46%), Chinese (28.64%), and Fast Food (20.79%), indicating that these cuisines dominate the restaurant landscape. The overlap in percentages suggests that many restaurants offer multiple cuisines.”

## TASK 3 : PRICE RANGE DESTRIBUTION 

1️⃣ Create a histogram or bar chart to visualize the distribution of restaurants across price ranges.

2️⃣ Calculate the percentage of restaurants in each price range category.

Essentially, we want to know:

* How many restaurants fall into each price category (1 = cheapest, 4 = most expensive).

* What percentage of the total restaurants belong to each category.

```python
# Check and handle missing values .

if 'price range' not in df.columns:
    raise KeyError("No 'price range' column found in the dataset.")
df['price range'].fillna(0, inplace=True)
```

```python
# Count number of restaurants in each price range
price_counts = df['price range'].value_counts().sort_index()
price_counts
```

#### Price range meaning usually:

* 1 → inexpensive

* 2 → moderate

* 3 → expensive

* 4 → very expensive

* value_counts() counts how many restaurants fall into each price range.

* sort_index() ensures the order is ascending (1, 2, 3, 4).

```python
# Calculate percentage of restaurants per price range

price_percent = (price_counts / len(df)) * 100
price_percent
```

```python
# Combine counts and percentages into one table

price_summary = pd.DataFrame({
    'Price Range': price_counts.index,
    'Count': price_counts.values,
    'Percentage': price_percent.round(2).values
})

print("Price Range Distribution:\n")
print(price_summary)
```

```python
# Visualization - Bar Chart .

plt.figure(figsize=(8,5))
plt.bar(price_summary['Price Range'], price_summary['Count'], color='skyblue', edgecolor='black')
plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range Category")
plt.ylabel("Number of Restaurants")
plt.xticks(price_summary['Price Range'])
plt.show()
```

```python
# Visualization - Percentage Bar Chart .

plt.figure(figsize=(8,5))
plt.bar(price_summary['Price Range'], price_summary['Percentage'], color='orange', edgecolor='black')
plt.title("Percentage of Restaurants by Price Range")
plt.xlabel("Price Range Category")
plt.ylabel("Percentage of Restaurants (%)")
plt.xticks(price_summary['Price Range'])
plt.show()
```

* Majority (~43%) of restaurants are low-cost (price range 1).

* Around 35% fall into mid-range (2).

* Only a small fraction (~6%) are high-end restaurants (4).

```python
# Pie plot (optional - For Understanding)

plt.figure(figsize=(7,7))
plt.pie(price_summary['Percentage'], 
        labels=price_summary['Price Range'],
        autopct='%1.1f%%', 
        startangle=140,
        wedgeprops={'edgecolor': 'black'})
plt.title("Percentage Distribution of Restaurants by Price Range")
plt.show()
```

### TASK 3 : FINAL SUMMARY 

Most restaurants belong to the lower price categories, indicating that the market is dominated by affordable dining options. High-end restaurants (price range 4) make up a small percentage of the total, showing a niche but limited premium market segment.

# TASK 4 : ONLINE DELIVERY 

1️⃣ Find what percentage of restaurants offer online delivery.

2️⃣ Compare the average rating between restaurants with and without online delivery.

In simple words — we’ll check:

* How common online delivery is, and

* Whether restaurants that offer it tend to have better or worse ratings.

```python
# Check and handle missing values

if 'has online delivery' not in df.columns:
    raise KeyError("No 'has online delivery' column found in the dataset.")
if 'aggregate rating' not in df.columns:
    raise KeyError("No 'aggregate rating' column found in the dataset.")
```

```python
df['has online delivery'].fillna('No', inplace=True)
df['aggregate rating'].fillna(0, inplace=True)

# If any restaurant doesn’t have info about online delivery → assume “No”.
# Missing ratings replaced with 0 to avoid math errors.
```

```python
# Calculate how many restaurants offer online delivery 

online_counts = df['has online delivery'].value_counts()
online_counts

# Counts how many restaurants say “Yes” or “No”.
```

```python
# Calculate percentage of restaurants offering online delivery

online_percent = (online_counts / len(df)) * 100
online_percent
```

```python
# Combine counts and percentages into one DataFrame.

online_summary = pd.DataFrame({
    'Online Delivery': online_counts.index,
    'Count': online_counts.values,
    'Percentage': online_percent.round(2).values
})

print("Online Delivery Availability:\n")
print(online_summary)
```

```python
# Compare average rating between delivery and non-delivery restaurants.

avg_rating_comparison = df.groupby('has online delivery')['aggregate rating'].mean().round(2).reset_index()
print("\nAverage Rating Comparison:\n")
print(avg_rating_comparison)
```

* Groups data into two categories: restaurants with and without online delivery.

* Finds average rating for each group.

* Helps you see whether online-delivery restaurants perform better in ratings.

```python
# Visualization 1 - Online Delivery Availability.

plt.figure(figsize=(7,5))
plt.bar(online_summary['Online Delivery'], online_summary['Percentage'], color=['skyblue', 'orange'])
plt.title("Percentage of Restaurants Offering Online Delivery")
plt.xlabel("Online Delivery Availability")
plt.ylabel("Percentage of Restaurants (%)")
plt.show()
```

```python
# Visualization 2 - Average Rating Comparison.

plt.figure(figsize=(7,5))
plt.bar(avg_rating_comparison['has online delivery'], avg_rating_comparison['aggregate rating'], color=['green', 'red'])
plt.title("Average Rating: With vs Without Online Delivery")
plt.xlabel("Online Delivery Availability")
plt.ylabel("Average Rating")
plt.show()
```

#### 1️⃣ Percentage of Restaurants Offering Online Delivery

Out of all restaurants in your dataset (~9,551 total),
7100 restaurants (≈74%) do not offer online delivery,
and 2451 restaurants (≈26%) do.

✅ Meaning:

* Only about 1 in 4 restaurants in this dataset provide online delivery.

💡 Possible real-world reasons:

* Many restaurants might be dine-in or local eateries that haven’t adopted delivery systems yet.

* This data could be slightly older, when online delivery (like Zomato, Swiggy) wasn’t universal.

#### 2️⃣ Average Ratings Comparison

| Delivery Type         | Average Rating | Interpretation                                       |
| --------------------- | -------------- | ---------------------------------------------------- |
| No Online Delivery    | **2.47**       | Lower ratings, possibly weaker customer satisfaction |
| Yes (Online Delivery) | **3.25**       | Higher ratings, better service reputation            |

Meaning:

* Restaurants offering online delivery have significantly higher average ratings (3.25 vs 2.47).

#### 1 . Analytical Insight:

* The analysis reveals that while only about 25.7% of restaurants provide online delivery, these restaurants maintain higher customer ratings on average (3.25 vs 2.47).
This suggests a correlation between online service availability and customer satisfaction or perception of quality.

#### 2 . Business Insight:

* Expanding online delivery could be a key strategy for restaurants to improve visibility, convenience, and overall ratings.
